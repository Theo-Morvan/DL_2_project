# RBM

```python
from src.rbm import RBM
from src.data import lire_alpha_digits, lire_alpha_digit_full
from src.utils import display_alpha_digits
%load_ext autoreload
%autoreload 2

X = lire_alpha_digits("data/binaryalphadigs.mat", "A")
display_alpha_digits(X)
```

```python
import matplotlib.pyplot as plt
```

```python
latent_dimensions = [20,80,110,140,200]
number_images = 5
fig, axes = plt.subplots(len(latent_dimensions), number_images, figsize=(30,40))
for i, element in enumerate(latent_dimensions):
    model = RBM(320, element)
    model.train(X, 0.1, 4, 100)
    generated_images = model.generer_image(20, number_images)
    for j in range(number_images):
        axes[i,j].imshow(generated_images[j].reshape((20,16)),cmap="gray")
        axes[i,j].set_xticks([])
        axes[i,j].set_yticks([])
plt.savefig("Images_A_RBM_Multiple_Layers")
```

```python
iterations_gibbs = [1,10,20,30,40]
number_images = 5
latent_dimension = 80
model = RBM(320, latent_dimension)
model.train(X, 0.1, 4, 300)

number_images = 5
fig, axes = plt.subplots(len(iterations_gibbs), number_images, figsize=(30,40))

for i, element in enumerate(iterations_gibbs):
    print(element)
    generated_images = model.generer_image(element, number_images)
    for j in range(number_images):
        axes[i,j].imshow(generated_images[j].reshape((20,16)),cmap="gray")
        axes[i,j].set_xticks([])
        axes[i,j].set_yticks([])
plt.savefig(f"Images_A_RBM_Variations_Gibbs_{latent_dimension}")
```

```python
generated = model.generer_image(10, 10)
display_alpha_digits(generated)
```

```python
X_ab = lire_alpha_digit_full(["a","b","c","d","e","f"])
latent_dimensions = [20,80,110,140,200]
number_images = 10
number_layers = 3
fig, axes = plt.subplots(len(latent_dimensions), number_images, figsize=(30,40))
for i, element in enumerate(latent_dimensions):
    model = DBN([320]+[element]*number_layers)
    model.train(X_ab, 0.1, 4, 100)
    generated_images = model.generer_image(30, number_images)
    for j in range(number_images):
        axes[i,j].imshow(generated_images[j].reshape((20,16)),cmap="gray")
        axes[i,j].set_xticks([])
        axes[i,j].set_yticks([])
        axes[i,j].axis("off")
plt.savefig("Images_A_DBN_Multiple_Layers_multiple_characters")
```

```python
latent_dimension = 80
number_images = 10
number_layers = [1,2,3,4,5]
fig, axes = plt.subplots(len(latent_dimensions), number_images, figsize=(30,20))
for i, element in enumerate(number_layers):
    model = DBN([320]+[latent_dimension]*element)
    model.train(X_ab, 0.1, 4, 2000)
    generated_images = model.generer_image(30, number_images)
    for j in range(number_images):
        axes[i,j].imshow(generated_images[j].reshape((20,16)),cmap="gray")
        axes[i,j].set_xticks([])
        axes[i,j].set_yticks([])
plt.savefig("Images_multiple_letters_DBN_impact_depth")
```
